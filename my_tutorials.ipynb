# METHODS TO NORMALIZING DATA

```python
# Simple feature scaling

df['depth'] = df['depth'] / df['depth'].max()
print(df['depth'])
```

```python
#Min - Max

df['depth'] = (df['depth'] / df['depth'].min()) / (df['depth'].max() - df['depth'].min())
print(df['depth'])
```
